```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
plt.rcParams['figure.figsize'] = 20, 6
plt.style.use('ggplot')
```

```python
n = 2000
lb, ub = 0, 100
x = np.linspace(lb, ub, n)
d1 = np.full((n,), 0.0)
d2 = np.full((n,), 5.0)
d3 = np.full((n,), 2.0)
y1 = np.sin(x+d1)
plt.plot(x, y1)
y2 = np.sin(x+d2)
plt.plot(x, y2)
y3 = np.sin(x+d3)
plt.plot(x, y3)
```

```python
k = 0.0013
for i in range(n):
    if i!=0:
        d1[i] = d1[i-1] + k*(np.sin(d2[i-1]-d1[i-1]) + np.sin(d3[i-1]-d1[i-1]))
        d2[i] = d2[i-1] + k*(np.sin(d3[i-1]-d2[i-1]) + np.sin(d1[i-1]-d2[i-1]))
        d3[i] = d3[i-1] + k*(np.sin(d1[i-1]-d3[i-1]) + np.sin(d2[i-1]-d3[i-1]))
y1 = np.sin(x+d1)
plt.plot(x, y1)
y2 = np.sin(x+d2)
plt.plot(x, y2)
y3 = np.sin(x+d3)
plt.plot(x, y3)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
for i in range(n):
    if i!=0:
        d1[i] = d1[i-1] + k*(np.sin(d2[i-1]-d1[i-1]) + np.sin(d3[i-1]-d1[i-1]))
        d2[i] = d2[i-1] + k*(np.sin(d3[i-1]-d2[i-1]) + np.sin(d1[i-1]-d2[i-1]))
        d3[i] = d3[i-1] + k*(np.sin(d1[i-1]-d3[i-1]) + np.sin(d2[i-1]-d3[i-1]))
        d1[i:] = d1[i]
        d2[i:] = d2[i]
        d3[i:] = d3[i]
    y1 = np.sin(x+d1)
    ax.plot(x, y1)
    y2 = np.sin(x+d2)
    ax.plot(x, y2)
    y3 = np.sin(x+d3)
    ax.plot(x, y3)
    p = i/20
    ax.set_xlim(p,p+20)
    ax.set_xticks([])
    fig.savefig('img/wave{0:04d}.jpg'.format(i))
    ax.cla()
fig.clf()
```
